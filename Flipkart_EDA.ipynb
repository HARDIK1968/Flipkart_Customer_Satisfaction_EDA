**We will first analyze Column named "channel_name" using Countplot.**

# **A countplot, a visualization tool within the Seaborn library, displays the counts of observations for each category of a categorical variable using bars, effectively acting as a histogram for categorical data. **

```python
plt.figure(figsize=(10, 5))

ax = sns.countplot(x='channel_name', data=df, palette="magma", order=df['channel_name'].value_counts().index)

plt.xticks(rotation=45, ha='right', fontsize=12, fontweight='bold')  # Bold bar names
plt.xlabel('Channel Name', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.title('Channel Name Counts', fontweight='bold')

for p in ax.patches:
    ax.annotate(f'{p.get_height()} ({p.get_height()/len(df)*100:.1f}%)',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, fontweight='bold', color='black', xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

***The terms Inbound, Outcall, and Email likely refer to different communication channels used in a customer service, sales, or call center environment:***

**Inbound  ** Represents incoming customer interactions, such as phone calls or messages initiated by customers seeking support, inquiries, or assistance.

**Outcall (Outbound Call) ** Refers to outgoing calls made by agents, often for sales, follow-ups, customer feedback, or promotional purposes.

**Email ** Represents customer interactions handled via email, including support queries, promotional campaigns, or transactional communications.

***In the given chart, Inbound is the most frequently used channel, followed by Outcall, while Email is the least used.***

**We will now Analyze Coumn named "Issue_reported at" and "issue_responded". The Datatypes of these Columns is "Object". We will have to Change the Datatype to "datetime64" for Further Analysis.**



---



```python
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], format='%d-%m-%Y %H:%M', errors='coerce')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], format='%d-%m-%Y %H:%M', errors='coerce')

df["Issue_resolution_time"] = df["issue_responded"] - df["Issue_reported at"]

print(df[['Issue_reported at', 'issue_responded', 'Issue_resolution_time']].head())
```

```python
df['Issue_resolution_time'].unique()
```

**The below code converts Issue_resolution_time (a timedelta column) into total hours, categorizes it into predefined resolution time bins, and assigns appropriate labels to each bin for analysis.**

```python
# Convert timedelta to total hours
df['Resolution_Hours'] = df['Issue_resolution_time'].dt.total_seconds() / 3600  # Convert seconds to hours

# Convert timedelta to total hours
df['Resolution_Hours'] = df['Issue_resolution_time'].dt.total_seconds() / 3600  # Convert seconds to hours

# Define bins (in hours) and corresponding labels
bins = [0, 1, 2, 3, 4, 5, 6, 12, 24, 48, 72, float('inf')]
labels = ['<1 hour', '<2 hours', '<3 hours', '<4 hours', '<5 hours', '<6 hours',
          '<12 hours', '>12 hours - <1 day', '>1 day - <2 days',
          '>2 days - <3 days', '>3 days']

# Apply binning
df['Resolution_Time_Bins'] = pd.cut(df['Resolution_Hours'], bins=bins, labels=labels, right=False)

# Display sample output
df[['Issue_resolution_time', 'Resolution_Hours', 'Resolution_Time_Bins']].head()
df['Resolution_Time_Bins'] = pd.cut(df['Resolution_Hours'], bins=bins, labels=labels, right=False)

# Display sample output
df[['Issue_resolution_time', 'Resolution_Hours', 'Resolution_Time_Bins']].head()
```

```python
plt.figure(figsize=(10, 5))

# Ensure categorical ordering
order = df['Resolution_Time_Bins'].value_counts().index

# Countplot with ordered bins
ax = sns.countplot(y='Resolution_Time_Bins', data=df, palette="viridis", order=order)

plt.xlabel("Number of Issues", fontsize=12, fontweight='bold')
plt.ylabel("Resolution Time Bin", fontsize=12, fontweight='bold')
plt.title("Distribution of Issues Across Resolution Time Bins", fontsize=14, fontweight='bold')

# Get total count of issues
total_count = df.shape[0]  # Correct way to get total rows

# Adjust x-axis limit to ensure labels fit well
plt.xlim(0, df['Resolution_Time_Bins'].value_counts().max() * 1.1)  # Add 10% extra space

# Add count and percentage labels
for p in ax.patches:
    width = p.get_width()
    percentage = (width / total_count) * 100  # Ensure correct division
    ax.annotate(f'{width:,.1f}\n({percentage:.1f}%)',
                (width, p.get_y() + p.get_height() / 2.),
                ha='left', va='center', fontsize=8, fontweight='bold', color='black',
                xytext=(5, 0), textcoords='offset points')

plt.tight_layout()
plt.show()
```

**Key Observations:**

1. Majority of issues (75.2%) were resolved in less than 1 hour (64,610 issues), as shown by the longest bar.

2. Other resolution time bins have significantly lower issue counts, with 4.6% of issues resolved within 2 hours and 3.7% within 12 hours to 1 day.

3. The percentage of issues declines as resolution time increases beyond 1 day.
Only 0.4% of issues took more than 3 days to resolve (313 issues).

***This chart highlights the efficiency of issue resolution, with most cases being resolved quickly.***

```python
# Replace missing values
df['Customer_City'] = df['Customer_City'].replace(["", "nan", "NaN", None], pd.NA)
df['Customer_City'] = df['Customer_City'].fillna("Not Available")

plt.figure(figsize=(10, 5))

# Select the top 10 categories
top_10_categories = df['Customer_City'].value_counts().sort_values(ascending=False).head(10).index

# Create a horizontal bar plot
ax = sns.countplot(y='Customer_City', data=df[df['Customer_City'].isin(top_10_categories)],
                   palette="magma", order=top_10_categories)

# Formatting
plt.yticks(fontsize=8, fontweight='bold')
plt.xlabel('Count', fontweight='bold')
plt.ylabel('Customer_City Count', fontweight='bold')
plt.title('Top 10 Customer_City Counts', fontweight='bold')

# Correct x-axis limit
plt.xlim(0, df['Customer_City'].value_counts().max() * 1.1)  # Add 10% extra space

# Adjusting label placement dynamically
for p in ax.patches:
    count = p.get_width()
    percentage = count / len(df) * 100

    # If bar is large, place text inside; otherwise, place outside
    if count > df['Customer_City'].value_counts().max() * 0.3:  # If >30% of max value
        text_x = count - 5000  # Adjust text inside
        ha_align = 'right'
        color = 'white'
    else:
        text_x = count + 500  # Adjust text outside
        ha_align = 'left'
        color = 'black'

    ax.annotate(f'{count:,.1f}\n({percentage:.1f}%)',
                (text_x, p.get_y() + p.get_height() / 2),
                ha=ha_align, va='center', fontsize=8, fontweight='bold', color=color)

plt.tight_layout()
plt.show()
```

**The image shows a horizontal bar chart visualizing the top 10 most frequent customer cities from a dataset. The key observations are:**

1. Not Available" dominates the dataset, with 68,828 entries (~80.1%), indicating missing or unrecorded city values.
2. The next most frequent cities, such as Hyderabad (722, 0.8%), New Delhi (688, 0.8%), and Pune (435, 0.5%), have significantly lower counts.
